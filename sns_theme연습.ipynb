# matplotlib의 각 객체에 접근하여 속성을 바꿔보도록 하자

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_context('talk')
sns.set_palette('Set2')
sns.set_style('whitegrid')

df = pd.read_csv('./Penguin/penguin.csv')

f,ax = plt.subplots(1,1,figsize=(18,6))

sns.scatterplot(x = df['bill_length_mm'], y = df['bill_depth_mm'], alpha=0.5, ax=ax)

plt.show()
```

```python
plt.rcParams['font.family']='AppleGothic'

f,ax = plt.subplots(1,2,figsize=(18,6),
                    gridspec_kw={'wspace':0.1},
                    constrained_layout=True)

x=[i for i in range(-5,6,1)]
y=[i**2 for i in x]
z=[i**3 for i in x]

ax[0].plot(x,y, marker='o', label='거듭제곱')
ax[1].plot(x,z, ls=':', label='세제곱')

for a in ax : 
    a.set_xlabel('x value')
    a.legend()
ax[0].set_ylabel('y value')

f.suptitle('CONFIGURATION PRACTICE')

plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

f,ax = plt.subplots(1,2,figsize=(18,6))

sns.violinplot(data=df, x='species',y='body_mass_g', hue='sex',split=True, ax=ax[0])
sns.violinplot(data=df, x='species',y='body_mass_g', hue='sex',split=True, ax=ax[1])

ax[0].set_ylim(ymin=2000)
ax[0].set(xlabel='test')
ax[0].spines[['left','top','right']].set_visible(False)
ax[0].tick_params(axis='y', length=10)

ax[1].set_ylim(ymin=2000)
ax[1].set(xlabel='test')
ax[1].spines[['left','top','right']].set_visible(False)
# ax[1].tick_params(axis='y', length=5)
ax[1].grid(axis='y', c='lightblue')
plt.tight_layout()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_style('whitegrid')

data = np.random.randint(10,size=10)

fig,ax = plt.subplots(1,2,figsize=(18,6))
fig.set_facecolor('white')
ax[0].plot(range(len(data)),data,marker='o')
sns.lineplot(x=range(len(data)),y=data,marker='o',ax=ax[1])
# width는 눈금폭
d = 'inout'
ax[0].tick_params(width=10, direction=d)
ax[1].tick_params(length=0, direction=d)
plt.tight_layout()
```

```python
# ax.tick_params(axis:str=변경할 축, length:float=눈금길이, width:float=눈금두께, direction:str=눈금방향, labelcolor:str=눈금숫자색상)
# direction : in, out, inout

import matplotlib.pyplot as plt

list_x_1 = [1, 2, 3, 4, 5, 6, 7, 8]
list_y_1 = [10, 2, 8, 4, 6, 5, 6, 4]


plt.plot(list_x_1, list_y_1,
         color='skyblue',
         marker='o', markerfacecolor='blue',
         markersize=6)
plt.tick_params(axis='y', length=8.8)
plt.tick_params(axis='x', labelcolor='red')
plt.show()
```
