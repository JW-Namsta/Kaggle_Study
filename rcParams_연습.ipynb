```python
import seaborn as sns
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd

df = pd.read_csv('train.csv')
sns.set_theme(font_scale=1.8, style='ticks', palette='viridis')
matplotlib.rcParams['axes.edgecolor']='lightblue'
matplotlib.rcParams['font.family']='AppleGothic' # 윈도우 OS의 경우 'Malgun Gothic'

f,ax = plt.subplots(1,2,figsize=(18,6))
sns.lineplot(data=df, x='Age', y='Survived', estimator=np.mean, ax=ax[0], errorbar=None)
sns.histplot(data=df, x='Age', ax=ax[1])
ax[0].set_title('한글테스트')

plt.tight_layout()
```
