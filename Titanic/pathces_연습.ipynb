```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
sns.set(font_scale=2.5)
import pandas as pd
import numpy as np

df_train = pd.read_csv('train.csv')
```

```python
# 막대그래프에 값 넣는거랑 bar 폭 조절하는 법인데 patches가 뭔지 알아야할 것 같음
f,ax = plt.subplots(1,2,figsize=(18,6))

##### PLOT - 1
sns.barplot(data=df_train[['Survived','Pclass']],x='Pclass',y='Survived',ax=ax[0], estimator=np.sum, errorbar=None,dodge=False)
print(ax[0].patches)
for p in ax[0].patches: 
    print(p)
    ##### annot

    ax[0].annotate("%.0f" % p.get_height(), (p.get_x() + p.get_width()/2., p.get_height() - 15), 
       ha='center', va='center', fontsize=10, color='black', xytext=(0, 10), 
       textcoords='offset points') 
    
    ##### bar width
    x = p.get_x()
    old_width = p.get_width()
    p.set_width(0.4)
    p.set_x(x+(old_width-0.4)/2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)
```

```python
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Distribution')

violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

violin['cbars'].set_edgecolor('gray')
violin['cmaxes'].set_edgecolor('red')
violin['cmins'].set_edgecolor('gray')
violin['cmeans'].set_edgecolor('gray')

# ax.add_patch(
#    patches.Rectangle(
#       (1.8, 1.0),                   # (x, y)
#       0.4, 1.5,                     # width, height
#       edgecolor = 'deeppink',
#       facecolor = 'lightgray',
#       fill=True,
#    ))

# ax.add_patch(
#      patches.Arrow(
#         2.0, 5.0, # 화살표 위치 x,y
#         0.0, -2, # 화살표 위치 변화량 dx,dy
#         width=0.3,
#         edgecolor = 'green',
#         facecolor = 'lightgreen'
#      ))

ax.annotate('local max', xy=(3, 1),  xycoords='data',
            xytext=(0.8, 0.95), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='top',
            )
            
ax.annotate("000", xy=(1,1))

ax.add_patch(
     patches.Polygon(
        ((1.9, 4.0), (2.0, 2.5), (2.1, 4.0)),
        closed=True,
        edgecolor = 'deeppink',
        facecolor = 'lightgray'
     ))

ax.text(1.5, 6, 'Max of Data B')

plt.show()
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

A = np.array([1,2,3,4,5])
B = np.array([1,15,20,10,15])

f,ax = plt.subplots(1,2,figsize=(18,6))

sns.pointplot(x=A,y=B,errorbar=None, ax=ax[0]) # ,markers=True, markersize=5
sns.barplot(x=A,y=B,errorbar=None, ax=ax[1])

# ax[0].annotate("%.0f" % p.get_height(), (p.get_x() + p.get_width()/2., p.get_height() - 15), 
#    ha='center', va='center', fontsize=10, color='black', xytext=(0, 10), 
#    textcoords='offset points') 
for dot in range(len(A)) :
    ax[0].annotate(
        text='{}'.format(B[dot]),
        xy=(A[dot]-1.2,B[dot]),
        fontsize=15,
        ha='center',
        va='center',)

print(ax[1].patches)
for bar in ax[1].patches : 
    print(bar, bar.get_height())
plt.show()
```
